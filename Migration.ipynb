```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create figure and axis
fig, ax = plt.subplots(figsize=(10, 12))

# Create rectangles
rect1 = patches.Rectangle((0.2, 0.85), 0.6, 0.1, edgecolor='black', facecolor='lightgray')
rect2 = patches.Rectangle((0.2, 0.7), 0.6, 0.1, edgecolor='black', facecolor='lightgray')
rect3 = patches.Rectangle((0.2, 0.55), 0.6, 0.1, edgecolor='black', facecolor='lightgray')
rect4 = patches.Rectangle((0.2, 0.4), 0.6, 0.1, edgecolor='black', facecolor='lightgray')
rect5 = patches.Rectangle((0.2, 0.25), 0.6, 0.1, edgecolor='black', facecolor='lightgray')
rect6 = patches.Rectangle((0.2, 0.1), 0.6, 0.1, edgecolor='black', facecolor='lightgray')

# Add rectangles to the plot
ax.add_patch(rect1)
ax.add_patch(rect2)
ax.add_patch(rect3)
ax.add_patch(rect4)
ax.add_patch(rect5)
ax.add_patch(rect6)

# Add arrows
for y_start, y_end in [(0.85, 0.75), (0.7, 0.6), (0.55, 0.45), (0.4, 0.3), (0.25, 0.15)]:
    ax.annotate('', xy=(0.5, y_end), xytext=(0.5, y_start), arrowprops=dict(facecolor='black', shrink=0.3))

# Add text to rectangles
ax.text(0.5, 0.9, 'Forced Migration and IDPs\n- Factors driving migration\n- Characteristics of IDPs', 
        horizontalalignment='center', verticalalignment='center', fontsize=11, color='black')
ax.text(0.5, 0.75, 'Demographic and Socio-economic Characteristics of IDPs\n- Age, gender, education, skills\n- Employment status\n- Socio-economic conditions', 
        horizontalalignment='center', verticalalignment='center', fontsize=11, color='black')
ax.text(0.5, 0.6, 'Impact on Labor Market\n- Employment rates\n- Job opportunities\n- Wage dynamics\n- Sector-specific impacts', 
        horizontalalignment='center', verticalalignment='center', fontsize=11, color='black')
ax.text(0.5, 0.45, 'Economic Contributions and Challenges\n- Positive contributions\n- Challenges', 
        horizontalalignment='center', verticalalignment='center', fontsize=11, color='black')
ax.text(0.5, 0.3, 'Mediating Factors\n- Social capital\n- Government policies\n- International aid', 
        horizontalalignment='center', verticalalignment='center', fontsize=11, color='black')
ax.text(0.5, 0.15, 'Overall Economic Impact\n- Balance of contributions and challenges', 
        horizontalalignment='center', verticalalignment='center', fontsize=11, color='black')

# Remove axes
ax.set_axis_off()

```
